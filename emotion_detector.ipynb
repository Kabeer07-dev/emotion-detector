```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('train.txt',sep=';',names=['text','emotion'])

df.head()
```

```python
unique_emotion = df['emotion'].unique()
unique_emotion
```

```python
emotion = {}

for i,em in enumerate(unique_emotion):
    emotion[em] = i

emotion
```

```python
df['emotion'] = df['emotion'].map(emotion)
df.head()
```

### Text cleaning and preprocessing

```python
df['text'] = df['text'].apply(lambda x:x.lower())
```

```python
# removing digits
def remove_digit(txt):
    new = ""
    for i in txt:
        if not i.isdigit():
            new += i

    return new


df['text'] = df['text'].apply(remove_digit)
```

```python
def remove_emoji(txt):
    new = ""
    for i in txt:
        if i.isascii():
            new += i

    return new

df['text'] = df['text'].apply(remove_emoji)
```

```python
import string

def remove_punc(txt):
    return txt.translate(str.maketrans('','',string.punctuation))


df['text'] = df['text'].apply(remove_punc)
```

```python
import nltk

nltk.download("stopwords")
```

```python
from nltk.corpus import stopwords

# Download once
# import nltk
# nltk.download("stopwords")

stop_words = set(stopwords.words("english"))

def remove_stopword(text):
    words = text.split()
    filtered_words = [word for word in words if word.lower() not in stop_words]
    return " ".join(filtered_words)
```

```python
df["text"] = df["text"].apply(remove_stopword)
```
